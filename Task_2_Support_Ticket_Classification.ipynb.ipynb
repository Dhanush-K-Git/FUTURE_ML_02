```python
import pandas as pd

# 1. Create the Dataset (Dictionary)
# We map typical keywords to categories manually for training
data = {
    'text': [
        "My internet is very slow and keeps buffering",
        "I cannot access the website, getting 404 error",
        "The connection drops frequently during calls",
        "My download speed is terrible today",
        "I was charged twice for the same month",
        "My bill is higher than expected, please explain",
        "I want to dispute a charge on my invoice",
        "Payment failed but money was deducted",
        "I forgot my password and cannot reset it",
        "Unable to login to my account app",
        "How do I change my profile settings?",
        "My account is locked, please help",
        "Router is blinking red light",
        "I need a refund for the last transaction",
        "System is not responding to my login attempts"
    ],
    'category': [
        "Technical", "Technical", "Technical", "Technical",
        "Billing",   "Billing",   "Billing",   "Billing",
        "Account",   "Account",   "Account",   "Account",
        "Technical", "Billing",   "Account"
    ]
}

# 2. Convert to a DataFrame (Table)
df = pd.DataFrame(data)

# 3. Show the first few rows to verify
print("âœ… Dataset Created Successfully!")
print(df.head())
print(f"\nTotal examples: {len(df)}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# 1. Create the "Pipeline" (The AI Machine)
# It has two parts:
# Part A: TfidfVectorizer (The Translator) -> Converts text to numbers
# Part B: MultinomialNB (The Brain) -> Classifies the numbers
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# 2. Train the Model
# We feed it the questions (X) and the answers (y) from our dataframe
model.fit(df['text'], df['category'])

# 3. Test it immediately!
# Let's ask it a question it has NEVER seen before
test_complaint = "I am facing lag issues and slow connectivity"
prediction = model.predict([test_complaint])

print("âœ… Model Trained Successfully!")
print(f"Test Complaint: '{test_complaint}'")
print(f"Predicted Department: {prediction[0]}")


```
